```python
# Chargement de dépendances

import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
# Discrétisation
A = 0
B = 500
N = 101  # Nombre de points de discrétisation
Delta = (B-A)/(N-1)
discretization_indexes = np.arange(N)
discretization = discretization_indexes*Delta
```

```python
# Paramètres du modèle
mu = -5
a = 50
sigma2 = 12
```

```python
# Données
observation_indexes = [0, 20, 40, 60, 80, 100]
depth = np.array([0, -4, -12.8, -1, -6.5, 0])
```

```python
# Indices des composantes correspondant aux observations et aux componsantes non observées
unknown_indexes = list(set(discretization_indexes)-set(observation_indexes))
```

## Question 2

On va noter $A=(Z(x_{j_{i}}))$ et $B=(Z(x)), x \notin [x_{j_{i}}]$.
Alors $f_{B|A=(z(x_{j_{i}}))} = \cfrac{exp(- \cfrac{1}{2} (b - E(B|A))^T CS_B^{-1} (b - E(B|A)))}{2 \pi \cfrac{N - n + 2}{2} \sqrt{det(CS_B)}} $
où $CS_B = C_B - C_{BA}C_A^{-1}C_{AB}$ et les matrices C sont les matrices de covariance des séries de données A et B.

# *Question informatique*

## Question 1

```python
def c(h, a, sig2):
    """Fonction qui calcule la covariance selon la distance. """
    try:
        return sig2*exp(-abs(h)/a)
    except ZeroDivisionError as identifier:
        print("La valeur de a doit être différente de 0.")


def mat_dis(dis):
    """Renvoyer une matrice de distance selon une liste de distance. """
    mat = np.zeros((len(dis), len(dis)))
    for i in range(len(dis)):
        for j in range(len(dis)):
            mat[i, j] = abs(dis[i] - dis[j])
    return mat


def mat_cov(dis, f, a, sig2):
    """
        Renvoyer une matrice de covariance.
        dis: matrice de distances
        f: fonction qui calcule la covariance
    """
    mat = np.zeros(dis.shape)
    for i in range(dis.shape[0]):
        for j in range(dis.shape[1]):
            mat[i, j] = f(dis[i, j], a, sig2)
    return mat
```

## Question 2

```python
mat_distance = mat_dis(discretization)
```

## Question 3

```python
mat_covariance = mat_cov(mat_distance, c, a, sigma2)
```

## Question 4

***Entre les observations***

```python
cov_22 = mat_covariance[observation_indexes, :]
cov_22 = cov_22[:, observation_indexes]
```

***Entre les inconnues***

```python
cov_11 = mat_covariance[unknown_indexes, :]
cov_11 = cov_11[:, unknown_indexes]
```

***Entre les observations et les inconnues***

```python
cov_21 = mat_covariance[observation_indexes, :]
cov_21 = cov_21[:, unknown_indexes]
cov_12 = cov_21.T
```

## Question 5

```python
# Espérance générale
esp = (np.ones(N)*mu).T
# Espérance conditionnelle
cov_22_inv = np.linalg.inv(cov_22)
esp_con = esp[unknown_indexes]+cov_12.dot(cov_22_inv).dot(depth.T-esp[observation_indexes])
```

## Question 6

```python
# Matrice de covariance
cov_cdt = cov_11-cov_12.dot(cov_22_inv).dot(cov_21)
# Diagonale de la matrice de covariance
cov_cdt_dgnl = [cov_cdt[i, i] for i in range(len(unknown_indexes))]

unknown_distance = np.array(unknown_indexes)*Delta
known_distance = np.array(observation_indexes)*Delta
```

```python
plt.plot(unknown_distance, cov_cdt_dgnl)
plt.title('Diagonale covariance')
plt.show()
```

Par conséquent, les distances inconnues ne sont pas décorrélées avec les distances connues : c'est logique car les distances connues imposent une contrainte de déformation du câble sous-marin.

## Question 7

```python
def fusion_2(c, unc, rel, sim):
    """
        Fusionner les résultats simulés et les observations.
        c: liste de distance des points connus
        unc: liste de distance des points inconnus
        rel: liste de profondeur des points connus
        sim: liste de profondeur simulée des points inconnus
    """
    dis = np.zeros(len(c)+len(unc))
    dep = np.zeros(len(c)+len(unc))
    for i in range(len(c)):
        dis[c[i]] = c[i]
        dep[c[i]] = rel[i]
    for i in range(len(unc)):
        dis[unc[i]] = unc[i]
        dep[unc[i]] = sim[i]
    return dis, dep
```

```python
# Fonction de simulation gaussienne du numpy
x = np.random.multivariate_normal(esp_con, cov_cdt).T
# Obtenir les résultats d'une simulation
sim = fusion_2(observation_indexes, unknown_indexes, depth, x)
plt.plot(sim[0]*Delta, sim[1])
plt.title('A single simulation')
plt.show()
```

Le fond marin est très irrégulier, ce qui reflète plutôt bien ce que l'on peut observer dans la nature.

## Question 8

```python
def longueur(depth, delta):
    """
        Calculer la longueur simulée.
        depth: liste de profondeur
        delta: distance entre chaque point
    """
    res = 0
    for i in range(len(depth)-1):
        res += sqrt(delta**2+(depth[i]-depth[i+1])**2)
    return res
```

## Question 9

```python
def simulation(esp, cov, delta, num=1):
    """
        Lancer des simulations, renvoyer les résultats.
        esp: liste d'espérance conditionnelle
        cov: matrice de convariance conditionnelle
        delta: distance entre chaque point
        num: nombre de simulation
    """
    x = np.random.multivariate_normal(esp, cov).T
    sim = fusion_2(observation_indexes, unknown_indexes, depth, x)
    res = sim[1]
    length = [longueur(sim[1], delta)]
    for i in range(num-1):
        x = np.random.multivariate_normal(esp, cov).T
        sim = fusion_2(observation_indexes, unknown_indexes, depth, x)
        res = np.vstack((res, sim[1]))
        length.append(longueur(sim[1], delta))
    return res, np.array(length)
```

```python
# Simulation 100
res = simulation(esp_con, cov_cdt, Delta, 100)
plt.plot(unknown_distance, esp_con)
plt.plot(discretization, np.mean(res[0], axis=0))
plt.title('100 simulations et espérance conditionnelle')
plt.show()
```

La moyenne varie beaucoup : cela est sûrement dû à l'écart entre les différentes simulations, et au fait que 100 simulations ne suffisent pas pour passer à l'approximation prévue dans la loi des grands nombres.

## Question 12

```python
long12 = []
for _ in range(100):
    swap = simulation(esp_con, cov_cdt, Delta, 1)
    long12.append(swap[1][0])
```

## Méthode 1 :
On approxime l'expérience par une loi gaussienne.

```python
#Cette fonction calcule l'esperance d'une liste de valeurs.
def esperance(X):
    s = 0
    N = len(X)
    for x in X:
        s += x
    return s / N

#Cette fonction calcule la variance d'une liste de valeurs.
def variance(X):
    X2 = [x**2 for x in X]
    return esperance(X2) - (esperance(X)**2)
```

```python
moy = esperance(long12)
sig = sqrt(variance(long12))
print(f" L'intervalle de confiance est [{moy - 2 * (sig / 10)}, {moy + 2 * (sig / 10)}]")
```

### Méthode 2 :
On part de la moyenne et on ajoute des valeurs jusqu'à atteindre 95% des valeurs dans notre intervalle.

```python
a, b = moy, moy
test = False
cpt = 0
while cpt < 95:
    a += -0.01
    b += 0.01
    cpt = 0
    for x in long12:
        if a <= x and x <= b:
            cpt += 1
print(f"L'intervalle de confiance est [{a}, {b}]")
```

L'intervalle de confiance de l'approximation gaussienne est bien inférieur au vrai intervalle de confiance. En effet, avec seulement 100 simulations, on ne peut pas encore clairement distinguer le comportement gaussien de cette expérience.
